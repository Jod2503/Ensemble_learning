```python
import pandas as pd
df=pd.read_csv('student_clustering.csv')
```

```python
df
```

```python
from sklearn.cluster import KMeans
```

```python
# find value of wcss  by curve
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(df)
    wcss.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
# n=4(elbow point)
```
