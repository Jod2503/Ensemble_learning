```python
import numpy as np
import pandas as pd
df=pd.read_csv('boston_house_prices.csv',header=None)
df
```

```python
df=df.drop(0)
```

```python
df.columns = df.iloc[0]
df = df[1:]
```

```python
df
```

```python
X=df.iloc[:,0:13]
y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
```

```python
reg1=LinearRegression()
reg2=SVR()
reg3=DecisionTreeRegressor()
```

```python
estimators=[('lr',reg1),('svr',reg2),('dt',reg3)]
from sklearn.ensemble import VotingRegressor
```

```python
for estimator in estimators:
    score=cross_val_score(estimator[1],X,y,cv=10,scoring='r2')
    print(score.mean())
```

```python
    vr=VotingRegressor(estimators=estimators)
    score=cross_val_score(vr,X,y,cv=10,scoring='r2')
    print(score.mean())
```
